```python
def cuentaPalabras(texto):
    listaPalabras = texto.split(' ')
    cuenta = len(listaPalabras)
    print("El número de palabras es:", cuenta)
    return cuenta

def cuentaPalabrasUnicas(texto = '', ignoraMayusculas = False):
    if(ignoraMayusculas):
        textoFinal = texto.lower()
    else:
        textoFinal = texto
    listaPalabras = textoFinal.split(' ')
    cuenta = len(set(listaPalabras))
    print("El número de palabras únicas es:", cuenta)
    return cuenta

def buscaPalabra(texto, palabra, ignoraMayusculas = False):
    if(ignoraMayusculas):
        textoFinal = texto.lower()
    else:
        textoFinal = texto
    listaPalabras = textoFinal.split(' ')
    cuenta = listaPalabras.count(palabra)
    print("La palabra", palabra, "aparece", cuenta, "veces")
    return cuenta

def creaDiccionarioPalabras(texto, ignoraMayusculas = False):
    if(ignoraMayusculas):
        textoFinal = texto.lower()
    else:
        textoFinal = texto
    
    diccionarioPalabras = {}
    listaPalabras = textoFinal.split(' ')
    setPalabras = set(listaPalabras)
    for palabra in setPalabras:
        cuenta = listaPalabras.count(palabra)
        diccionarioPalabras.update({palabra : cuenta})
    return diccionarioPalabras

def borraPuntuacion(texto):
    textoFinal = texto.replace('\n', '').replace(';', '').replace(',', '').replace('.', '').replace('?', '')
    return textoFinal

f = open("8800.txt", "r")
texto = f.read()

#Extrae los primeros 5000 caracteres
textoLimitado = texto[0:5000]
print("Vamos a realizar análisis dentro de los primeros 5000 caracteres")

#Elimina los signos de puntuación
textoLimpio = borraPuntuacion(textoLimitado)

#Cuenta las palabras
cuentaPalabras(textoLimpio)

#Cuenta las palabras únicas
cuentaPalabrasUnicas(textoLimpio)

#Cuenta las palabras únicas quitando mayúsculas
cuentaPalabrasUnicas(textoLimpio, True)

#¿Cuántas veces aparece la palabra the en ese rango?
#¿Puedo saber cuantas veces existe una palabra en el texto cargado?
buscaPalabra(textoLimpio, 'the', True)

#Crea un diccionario donde asignes el número de apariciones a la palabra existente
creaDiccionarioPalabras(textoLimpio, True)
```
